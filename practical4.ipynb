```python
ytrain = pd.to_datetime(ytrain, errors='coerce')

# Convert to number (days since reference date)
ytrain = (ytrain - ytrain.min()).dt.days
```

```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.impute import SimpleImputer

# Step 1: Ensure xtrain is DataFrame
xtrain = pd.DataFrame(xtrain)

# Step 2: Convert categorical → numeric
xtrain = pd.get_dummies(xtrain)

# Step 3: Replace inf with NaN
xtrain = xtrain.replace([np.inf, -np.inf], np.nan)

# Step 4: Handle missing values
imputer = SimpleImputer(strategy="mean")
xtrain = imputer.fit_transform(xtrain)

# Step 5: Fix ytrain
ytrain = np.nan_to_num(ytrain)

# Step 6: Train model
lm = LinearRegression()
model = lm.fit(xtrain, ytrain)

print("Model training completed")
```

```python
from sklearn.metrics import r2_score

ypred = model.predict(xtrain)

print("R2 Score:", r2_score(ytrain, ypred))
```

```python
print(model.coef_)
print(model.intercept_)
```

```python
import matplotlib.pyplot as plt

plt.scatter(ytrain, ypred)
plt.xlabel("Actual")
plt.ylabel("Predicted")
plt.title("Actual vs Predicted")
plt.show()
```
