```python
import pandas as pd
import numpy as np

# Load CSV
df = pd.read_csv("STUDENT PERFORMANCE1.CSV")

# Calculate quartiles
q1 = np.percentile(df['math score'], 35)
q3 = np.percentile(df['math score'], 78)

# IQR
IQR = q3 - q1

# Print output
print("Inter Quartile Range (IQR) =", IQR)
```

```python
lower_bound=q1-1.5*IQR
print("Lower Bound=",lower_bound)
```
